```python
import cv2
```

```python
##webcam
cam=cv2.VideoCapture(0)
while True:
    ret,frame=cam.read()  # storing image in frame
    cv2.imshow("webcam",frame) ##displaying video
    if cv2.waitKey(1)==13:   ##ASCII 13 for enter key
        break
cam.release()
cv2.destroyAllWindows()
    
```
